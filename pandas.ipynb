```python
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data.csv')

ids = data['Responder_id']
responses = data['LanguagesWorkedWith']

count_language = Counter()  #  count the languages per person

for index in responses:
    count_language.update(index.split(';'))

# index.split(';')

languages = []
popularity = []

for item in count_language.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

#  reverse the data to ascesnding order
languages.reverse()
popularity.reverse()

# plot the data in bar chart, switch it horinzontally and reverse the X & Y axis
plt.barh(languages, popularity)
plt.title('Most Popula Languages Used By The Programmers According Stack-Over-Flow Survey')
plt.xlabel('Programming Languages')
plt.ylabel('Number of poeple that use them')
plt.grid(axis='x')
plt.savefig('programming_language_popularity.png', bbox_inches='tight')
plt.show()
```
