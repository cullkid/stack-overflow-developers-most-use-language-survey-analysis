```python
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data.csv')

ids = data['Responder_id']
responses = data['LanguagesWorkedWith']

count_language = Counter()  #  count the languages per person

for index in responses:
    count_language.update(index.split(';'))

# index.split(';')

languages = []
popularity = []

for item in count_language.most_common(5):
    languages.append(item[0])
    popularity.append(item[1])

languages
popularity

colors = ['#FFBDBD','#A8F09C','#FFE6BD', '#D9DAA0','#B2FCE4']
explode = [0.1,0,0,0,0]

plt.pie(popularity, labels=languages, colors=colors, wedgeprops={'edgecolor': 'black'},
        startangle=90, explode=explode, shadow=True, autopct='%1.1f%%')
plt.title('Most Popula Languages Used By The Programmers According Stack-Over-Flow Survey')
plt.tight_layout()
plt.savefig('Pie_chart_programming_language_popularity.png', bbox_inches='tight')
plt.show()
```
